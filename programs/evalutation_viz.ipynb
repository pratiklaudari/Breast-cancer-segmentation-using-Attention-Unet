```python
import csv
import matplotlib.pyplot as plt
import numpy as np
file_path=r"C:\Users\sapko\Downloads\segmentation_eval.csv"
```

```python
e=[]
d_c=[]
j_c=[]
l=[]
a=[]
v_l=[]
v_a=[]
```

```python
with open(file_path, 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        e.append(float(row[0]))
        d_c.append(float(row[1]))
        j_c.append(float(row[2]))
        l.append(float(row[3]))
        a.append(float(row[4]))
        v_l.append(float(row[5]))
        v_a.append(float(row[6]))
```

```python
plt.plot(e, d_c)
plt.xlabel("epoch")
plt.ylabel("dice")
plt.title("dice coefficient")
plt.show()
```

```python
plt.plot(e, j_c)
plt.xlabel("epoch")
plt.ylabel("jaccard")
plt.title("jaccard coefficient")
plt.show()
```

```python
plt.plot(e, l)
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("loss")
plt.show()
```

```python
plt.plot(e, a)
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.title("accuracy")
plt.show()
```

```python
plt.plot(e, v_l)
plt.xlabel("epoch")
plt.ylabel("val_loss")
plt.title("validation loss")
plt.show()
```

```python
plt.plot(e, v_a)
plt.xlabel("epoch")
plt.ylabel("val_accuracy")
plt.title("validation accuracy")
plt.show()
```
